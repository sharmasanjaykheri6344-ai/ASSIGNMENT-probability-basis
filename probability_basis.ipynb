Q.10  Differentiate between discrete and continuous random variables with examples.

..> Discrete vs Continuous Random Variables
1. Discrete Random Variable

A random variable that takes countable values
(either finite or infinite countable).

Key Features

Values are distinct and separate

Gaps exist between numbers

Probability of each value can be listed

Examples

Number of students in a class (e.g., 30, 31, 32)

Number of heads when flipping 3 coins

Number of cars passing a signal in 1 minute

Marks in an exam (if given as whole numbers)

2. Continuous Random Variable

A random variable that takes any value in a given interval
(infinite and uncountable).

Key Features

Values form a continuum

No gaps between values

Probabilities are assigned over intervals, not individual points

Examples

Height of students (e.g., 162.4 cm, 162.45 cm)

Time taken to complete a task

Temperature at a place

Weight of a person

Summary Table
Feature	Discrete RV	Continuous RV
Number of possible values	Countable	Infinite, uncountable
Gaps between values	Yes	No
Probability assigned to single value	Possible	Always 0
Examples	Number of goals, number of calls	Height, weight, time.
